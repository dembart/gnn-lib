```python
from ase.io import read
from ase.optimize import FIRE
from gnn_lib.calculators import CGCNNCalculator
```

```python
atoms = read('./data/Li10Ge(PS6)2_mp.cif')
```

```python
atoms
```

```python
# This models poorly predicts forces and stress, use it with caution!

calc = CGCNNCalculator(model='matpes_pbe_small', device='cuda')

atoms.calc = calc
```

```python
# energy per atom
atoms.get_potential_energy() / len(atoms)
```

```python
# maximum force component
abs(atoms.get_forces()).max()
```

```python
# stress tensor
from ase.units import GPa
atoms.get_stress() * GPa
```

```python
optim = FIRE(atoms)
optim.run(fmax=0.05)
```

```python
from ase.io import write
write('lgps_test_cgcnn_optim.cif', atoms)
```
